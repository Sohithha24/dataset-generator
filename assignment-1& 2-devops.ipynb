```python
# ========================================
# STEP 2: VERIFY DATASET SIZES
# ========================================

def estimate_dataset_size(dataset, name, sample_size=1000):
    """Estimate the size of a dataset in GB"""
    
    # Get actual sample size
    sample_size = min(sample_size, len(dataset))
    
    # Calculate total characters in sample
    total_chars = 0
    for i in range(sample_size):
        if 'text' in dataset[i]:
            total_chars += len(dataset[i]['text'])
        elif 'content' in dataset[i]:
            total_chars += len(dataset[i]['content'])
    
    # Estimate total size
    avg_chars_per_item = total_chars / sample_size
    total_chars_estimate = avg_chars_per_item * len(dataset)
    
    # Convert to MB/GB (1 char ≈ 1 byte in UTF-8 for English)
    size_mb = total_chars_estimate / (1024 * 1024)
    size_gb = size_mb / 1024
    
    print(f" {name}:")
    print(f"   Items: {len(dataset):,}")
    print(f"   Avg chars/item: {avg_chars_per_item:,.0f}")
    print(f"   Estimated size: {size_mb:.1f} MB ({size_gb:.3f} GB)")
    print()
    
    return size_gb

print("=" * 60)
print("CHECKING DATASET SIZES")
print("=" * 60)
print()

# Check Wikipedia
wiki_size = estimate_dataset_size(wiki, "Wikipedia")

# Check C4 (it's a list, need to wrap it)
class SimpleDataset:
    def __init__(self, data_list):
        self.data = data_list
    def __len__(self):
        return len(self.data)
    def __getitem__(self, idx):
        return self.data[idx]

c4_obj = SimpleDataset(c4_samples)
c4_size = estimate_dataset_size(c4_obj, "C4")

# Check OpenWebText
owt_size = estimate_dataset_size(owt, "OpenWebText")

# Total
total_size = wiki_size + c4_size + owt_size
print("=" * 60)
print(f"TOTAL DATASET SIZE: {total_size:.3f} GB")
print(f"   ({total_size * 1024:.0f} MB)")
print()
if total_size >= 1.0:
    print(" MEETS 1GB REQUIREMENT!")
    print(" Assignment requirement satisfied!")
else:
    print(" Below 1GB - need more data")
print("=" * 60)
```

```python
# ========================================
# STEP 3: DATA CLEANING PIPELINE
# ========================================

import re
from collections import defaultdict

print(" STARTING DATA CLEANING PIPELINE")
print("=" * 60)

def clean_text(text):
    """Clean and normalize a single text document"""
    if not text or len(text.strip()) == 0:
        return None
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text)
    
    # Remove special characters but keep basic punctuation
    text = re.sub(r'[^\w\s.,!?;:\-\'\"()]', '', text)
    
    # Strip leading/trailing whitespace
    text = text.strip()
    
    # Convert to lowercase for normalization
    text = text.lower()
    
    return text

def is_high_quality(text, min_words=50):
    """Check if document meets quality standards"""
    if not text:
        return False
    
    # Count words
    words = text.split()
    if len(words) < min_words:
        return False
    
    # Check if it's mostly text (not just numbers/symbols)
    alpha_chars = sum(c.isalpha() for c in text)
    if alpha_chars < len(text) * 0.5:  # At least 50% letters
        return False
    
    return True

def remove_duplicates(texts):
    """Remove duplicate documents"""
    seen = set()
    unique_texts = []
    
    for text in texts:
        # Use first 200 chars as fingerprint
        fingerprint = text[:200] if len(text) > 200 else text
        if fingerprint not in seen:
            seen.add(fingerprint)
            unique_texts.append(text)
    
    return unique_texts

print("\n1️ Cleaning Wikipedia...")
wiki_texts = []
for item in wiki:
    text = clean_text(item['text'])
    if text and is_high_quality(text):
        wiki_texts.append(text)

print(f"   Original: {len(wiki):,} articles")
print(f"   After cleaning: {len(wiki_texts):,} articles")
print(f"   Removed: {len(wiki) - len(wiki_texts):,} low-quality")

# Remove duplicates
wiki_texts = remove_duplicates(wiki_texts)
print(f"   After deduplication: {len(wiki_texts):,} unique articles")

print("\n2️ Cleaning C4...")
c4_texts = []
for item in c4_samples:
    text = clean_text(item['text'])
    if text and is_high_quality(text):
        c4_texts.append(text)

print(f"   Original: {len(c4_samples):,} samples")
print(f"   After cleaning: {len(c4_texts):,} samples")
print(f"   Removed: {len(c4_samples) - len(c4_texts):,} low-quality")

c4_texts = remove_duplicates(c4_texts)
print(f"   After deduplication: {len(c4_texts):,} unique samples")

print("\n3️ Cleaning OpenWebText...")
owt_texts = []
for item in owt:
    text = clean_text(item['text'])
    if text and is_high_quality(text):
        owt_texts.append(text)

print(f"   Original: {len(owt):,} samples")
print(f"   After cleaning: {len(owt_texts):,} samples")
print(f"   Removed: {len(owt) - len(owt_texts):,} low-quality")

owt_texts = remove_duplicates(owt_texts)
print(f"   After deduplication: {len(owt_texts):,} unique samples")

# Combine all cleaned texts
all_texts = wiki_texts + c4_texts + owt_texts

print("\n" + "=" * 60)
print(" CLEANING COMPLETE!")
print("=" * 60)
print(f"\n Final Statistics:")
print(f"   Wikipedia: {len(wiki_texts):,} documents")
print(f"   C4: {len(c4_texts):,} documents")
print(f"   OpenWebText: {len(owt_texts):,} documents")
print(f"   TOTAL CLEANED: {len(all_texts):,} documents")
print(f"\n Ready for tokenization!")
```

```python
# ========================================
# STEP 4: TOKENIZATION
# ========================================

from transformers import AutoTokenizer
import numpy as np

print("STARTING TOKENIZATION")
print("=" * 60)

# Load a pre-trained tokenizer (GPT-2 style, BPE)
print("\n1️ Loading tokenizer...")
print("   Using GPT-2 tokenizer (BPE - Byte Pair Encoding)")
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# Set padding token (GPT-2 doesn't have one by default)
tokenizer.pad_token = tokenizer.eos_token

print(f" Tokenizer loaded!")
print(f"   Vocabulary size: {tokenizer.vocab_size:,}")
print(f"   Model max length: {tokenizer.model_max_length:,}")

# Set our block size (max sequence length)
block_size = 512  # Common for training
print(f"   Our block size: {block_size}")

print("\n2️ Tokenizing all documents...")
print("   (This takes 2-3 minutes...)")

all_token_ids = []
docs_processed = 0
docs_too_short = 0

for text in all_texts:
    # Tokenize
    tokens = tokenizer.encode(text, add_special_tokens=True)
    
    # Skip very short documents (less than 10 tokens)
    if len(tokens) < 10:
        docs_too_short += 1
        continue
    
    # Handle long sequences: split into chunks of block_size
    for i in range(0, len(tokens), block_size):
        chunk = tokens[i:i + block_size]
        
        # Only keep chunks with at least 50 tokens
        if len(chunk) >= 50:
            # Pad if necessary
            if len(chunk) < block_size:
                chunk = chunk + [tokenizer.pad_token_id] * (block_size - len(chunk))
            
            all_token_ids.append(chunk)
    
    docs_processed += 1
    
    # Progress update
    if docs_processed % 10000 == 0:
        print(f"   Processed {docs_processed:,} / {len(all_texts):,} documents...")

print(f"\n Tokenization complete!")
print(f"   Documents processed: {docs_processed:,}")
print(f"   Documents skipped (too short): {docs_too_short:,}")
print(f"   Total token sequences created: {len(all_token_ids):,}")
print(f"   Each sequence length: {block_size} tokens")

# Convert to numpy array for efficiency
print("\n3️ Converting to numpy array...")
tokenized_data = np.array(all_token_ids, dtype=np.int32)
print(f" Shape: {tokenized_data.shape}")
print(f"   ({tokenized_data.shape[0]:,} sequences × {tokenized_data.shape[1]} tokens)")

# Calculate statistics
total_tokens = tokenized_data.shape[0] * tokenized_data.shape[1]
print(f"\n Tokenization Statistics:")
print(f"   Total tokens: {total_tokens:,}")
print(f"   Vocabulary size: {tokenizer.vocab_size:,}")
print(f"   Block size: {block_size}")
print(f"   Sequences: {len(all_token_ids):,}")

# Show sample
print(f"\n Sample tokenized sequence (first 20 tokens):")
print(f"   Token IDs: {all_token_ids[0][:20]}")
print(f"   Decoded: {tokenizer.decode(all_token_ids[0][:20])}")

print("\n" + "=" * 60)
print(" TOKENIZATION COMPLETE!")
print(" Ready for data loader!")
print("=" * 60)
```

```python
# ========================================
# STEP 5: PYTORCH DATA LOADER
# ========================================

import torch
from torch.utils.data import Dataset, DataLoader

print(" CREATING PYTORCH DATA LOADER")
print("=" * 60)

# Custom Dataset class
class TokenizedTextDataset(Dataset):
    """Custom PyTorch Dataset for tokenized text data"""
    
    def __init__(self, token_ids):
        """
        Args:
            token_ids: numpy array of shape (num_sequences, block_size)
        """
        self.token_ids = torch.tensor(token_ids, dtype=torch.long)
    
    def __len__(self):
        """Return the number of sequences"""
        return len(self.token_ids)
    
    def __getitem__(self, idx):
        """
        Get a single sequence
        Returns:
            Dictionary with input_ids and labels (for language modeling)
        """
        sequence = self.token_ids[idx]
        
        # For language modeling: input = sequence, target = sequence shifted by 1
        return {
            'input_ids': sequence,
            'labels': sequence  # In real training, you'd shift this by 1
        }

print("\n1️ Creating PyTorch Dataset...")
dataset = TokenizedTextDataset(tokenized_data)
print(f" Dataset created!")
print(f"   Total sequences: {len(dataset):,}")
print(f"   Sequence length: {dataset.token_ids.shape[1]}")

print("\n2️ Creating DataLoader...")
# Create DataLoader with batching and shuffling
batch_size = 32  # Common batch size for transformer training
num_workers = 0  # Use 0 for Kaggle/Colab compatibility

train_loader = DataLoader(
    dataset,
    batch_size=batch_size,
    shuffle=True,  # Shuffle for training
    num_workers=num_workers,
    pin_memory=True  # Faster data transfer to GPU
)

print(f" DataLoader created!")
print(f"   Batch size: {batch_size}")
print(f"   Number of batches: {len(train_loader):,}")
print(f"   Shuffle: True")
print(f"   Total samples per epoch: {len(dataset):,}")

print("\n3️ Testing DataLoader...")
# Get a sample batch
sample_batch = next(iter(train_loader))
print(f" Sample batch loaded!")
print(f"   Batch shape: {sample_batch['input_ids'].shape}")
print(f"   (batch_size={sample_batch['input_ids'].shape[0]}, sequence_length={sample_batch['input_ids'].shape[1]})")

# Decode first sequence in batch to show it's real text
first_sequence = sample_batch['input_ids'][0].tolist()
decoded_text = tokenizer.decode(first_sequence[:100])  # First 100 tokens
print(f"\n Sample from first batch:")
print(f"   {decoded_text}...")

print("\n" + "=" * 60)
print(" DATA LOADER COMPLETE!")
print("=" * 60)

print(f"\n Final Pipeline Summary:")
print(f"    Raw datasets: 3 sources (Wikipedia, C4, OpenWebText)")
print(f"    Total raw text: 1.4 GB")
print(f"    Cleaned documents: {len(all_texts):,}")
print(f"    Tokenized sequences: {len(dataset):,}")
print(f"    Total tokens: {len(dataset) * 512:,}")
print(f"    Batch size: {batch_size}")
print(f"    Batches per epoch: {len(train_loader):,}")
print(f"\n Ready for model training!")
```

```python
# ========================================
# ASSIGNMENT 2: MINI-GPT TRANSFORMER MODEL
# ========================================
import torch
import torch.nn as nn
import torch.nn.functional as F
import math

print("CREATING IMPROVED MODEL WITH DROPOUT")
print("=" * 60)

# Define hyperparameters (in case they were lost)
VOCAB_SIZE = tokenizer.vocab_size  # 50257
EMBED_DIM = 128
NUM_HEADS = 4
NUM_LAYERS = 2
MAX_SEQ_LEN = 512
FF_DIM = EMBED_DIM * 4  # 512
DROPOUT_RATE = 0.2

print(f"\nHyperparameters:")
print(f"   Vocab size: {VOCAB_SIZE:,}")
print(f"   Embed dim: {EMBED_DIM}")
print(f"   Num heads: {NUM_HEADS}")
print(f"   Num layers: {NUM_LAYERS}")
print(f"   Dropout: {DROPOUT_RATE}")

# Improved model classes with dropout

class MultiHeadAttentionWithDropout(nn.Module):
    def __init__(self, embed_dim, num_heads, dropout=0.1):
        super().__init__()
        assert embed_dim % num_heads == 0
        
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads
        
        self.q_proj = nn.Linear(embed_dim, embed_dim)
        self.k_proj = nn.Linear(embed_dim, embed_dim)
        self.v_proj = nn.Linear(embed_dim, embed_dim)
        self.out_proj = nn.Linear(embed_dim, embed_dim)
        
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        batch_size, seq_len, embed_dim = x.shape
        
        Q = self.q_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        K = self.k_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        V = self.v_proj(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        
        scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(self.head_dim)
        mask = torch.triu(torch.ones(seq_len, seq_len), diagonal=1).bool().to(x.device)
        scores = scores.masked_fill(mask, float('-inf'))
        
        attn_weights = F.softmax(scores, dim=-1)
        attn_weights = self.dropout(attn_weights)
        
        attn_output = torch.matmul(attn_weights, V)
        attn_output = attn_output.transpose(1, 2).contiguous().view(batch_size, seq_len, embed_dim)
        output = self.out_proj(attn_output)
        
        return output

class FeedForwardWithDropout(nn.Module):
    def __init__(self, embed_dim, ff_dim, dropout=0.1):
        super().__init__()
        self.fc1 = nn.Linear(embed_dim, ff_dim)
        self.fc2 = nn.Linear(ff_dim, embed_dim)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

class TransformerBlockWithDropout(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout=0.1):
        super().__init__()
        self.attention = MultiHeadAttentionWithDropout(embed_dim, num_heads, dropout)
        self.ff = FeedForwardWithDropout(embed_dim, ff_dim, dropout)
        self.ln1 = nn.LayerNorm(embed_dim)
        self.ln2 = nn.LayerNorm(embed_dim)
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        x = x + self.dropout(self.attention(self.ln1(x)))
        x = x + self.dropout(self.ff(self.ln2(x)))
        return x

class ImprovedMiniGPT(nn.Module):
    def __init__(self, vocab_size, embed_dim, num_heads, num_layers, max_seq_len, ff_dim=None, dropout=0.1):
        super().__init__()
        
        if ff_dim is None:
            ff_dim = embed_dim * 4
        
        self.token_embedding = nn.Embedding(vocab_size, embed_dim)
        self.position_embedding = nn.Embedding(max_seq_len, embed_dim)
        self.dropout = nn.Dropout(dropout)
        
        self.blocks = nn.ModuleList([
            TransformerBlockWithDropout(embed_dim, num_heads, ff_dim, dropout)
            for _ in range(num_layers)
        ])
        
        self.ln_final = nn.LayerNorm(embed_dim)
        self.head = nn.Linear(embed_dim, vocab_size)
        
        self.apply(self._init_weights)
        
    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                torch.nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            torch.nn.init.normal_(module.weight, mean=0.0, std=0.02)
    
    def forward(self, input_ids):
        batch_size, seq_len = input_ids.shape
        
        token_emb = self.token_embedding(input_ids)
        positions = torch.arange(0, seq_len, device=input_ids.device).unsqueeze(0)
        pos_emb = self.position_embedding(positions)
        
        x = self.dropout(token_emb + pos_emb)
        
        for block in self.blocks:
            x = block(x)
        
        x = self.ln_final(x)
        logits = self.head(x)
        
        return logits

# Clear GPU memory and set device
torch.cuda.empty_cache()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Create improved model
model_v2 = ImprovedMiniGPT(
    vocab_size=VOCAB_SIZE,
    embed_dim=EMBED_DIM,
    num_heads=NUM_HEADS,
    num_layers=NUM_LAYERS,
    max_seq_len=MAX_SEQ_LEN,
    ff_dim=FF_DIM,
    dropout=DROPOUT_RATE
).to(device)

# Count parameters
total_params_v2 = sum(p.numel() for p in model_v2.parameters())

print(f"\nImproved Model Created:")
print(f"   Dropout rate: {DROPOUT_RATE}")
print(f"   Total parameters: {total_params_v2:,}")
print(f"   Device: {device}")
print("\nReady to train improved model!")
```

```python
# ========================================
# TRAINING LOOP FOR MINI-GPT
# ========================================


import torch.optim as optim
from tqdm import tqdm
import time

print("TRAINING IMPROVED MODEL WITH DROPOUT")
print("=" * 60)

# Training configuration
LEARNING_RATE = 5e-4
NUM_EPOCHS_V2 = 2  # Fewer epochs to prevent overfitting
GRAD_CLIP = 1.0
BATCHES_PER_EPOCH = 1000
BATCH_SIZE = 16  # Same as before

print(f"\nConfiguration:")
print(f"   Learning rate: {LEARNING_RATE}")
print(f"   Epochs: {NUM_EPOCHS_V2} (reduced to prevent overfitting)")
print(f"   Dropout: {DROPOUT_RATE}")
print(f"   Batches per epoch: {BATCHES_PER_EPOCH}")
print(f"   Batch size: {BATCH_SIZE}")

# Create DataLoader (in case it was lost)
from torch.utils.data import DataLoader

train_loader_v2 = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True, pin_memory=True)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer_v2 = optim.Adam(model_v2.parameters(), lr=LEARNING_RATE)

# Storage
train_losses_v2 = []
perplexities_v2 = []

print("\nSTARTING TRAINING")
print("=" * 60)

start_time = time.time()

for epoch in range(NUM_EPOCHS_V2):
    print(f"\nEPOCH {epoch+1}/{NUM_EPOCHS_V2}")
    print("=" * 60)
    
    model_v2.train()
    epoch_loss = 0
    batch_count = 0
    
    pbar = tqdm(train_loader_v2, desc=f"Epoch {epoch+1}/{NUM_EPOCHS_V2}", total=BATCHES_PER_EPOCH)
    
    for batch in pbar:
        input_ids = batch['input_ids'].to(device)
        labels = batch['labels'].to(device)
        
        # Forward pass
        logits = model_v2(input_ids)
        
        # Calculate loss
        logits_flat = logits.view(-1, logits.size(-1))
        labels_flat = labels.view(-1)
        loss = criterion(logits_flat, labels_flat)
        
        # Backward pass
        optimizer_v2.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model_v2.parameters(), GRAD_CLIP)
        optimizer_v2.step()
        
        # Track metrics
        epoch_loss += loss.item()
        batch_count += 1
        pbar.set_postfix({'loss': f'{loss.item():.4f}'})
        
        # Stop after 1000 batches
        if batch_count >= BATCHES_PER_EPOCH:
            break
        
        # Clear cache periodically
        if batch_count % 100 == 0:
            torch.cuda.empty_cache()
    
    # Calculate epoch metrics
    avg_loss = epoch_loss / batch_count
    perplexity = torch.exp(torch.tensor(avg_loss)).item()
    
    train_losses_v2.append(avg_loss)
    perplexities_v2.append(perplexity)
    
    print(f"\nEpoch {epoch+1} Complete:")
    print(f"   Loss: {avg_loss:.4f}")
    print(f"   Perplexity: {perplexity:.2f}")

end_time = time.time()
training_time = end_time - start_time

print("\n" + "=" * 60)
print("TRAINING COMPLETE!")
print("=" * 60)
print(f"\nTraining time: {training_time/60:.1f} minutes")
print(f"\nFinal Results:")
print(f"   Initial Loss: {train_losses_v2[0]:.4f} -> Final Loss: {train_losses_v2[-1]:.4f}")
print(f"   Initial Perplexity: {perplexities_v2[0]:.2f} -> Final Perplexity: {perplexities_v2[-1]:.2f}")
print(f"   Improvement: {((train_losses_v2[0] - train_losses_v2[-1])/train_losses_v2[0]*100):.1f}%")

# Save checkpoint
print("\nSaving improved model checkpoint...")
torch.save({
    'model_state_dict': model_v2.state_dict(),
    'optimizer_state_dict': optimizer_v2.state_dict(),
    'train_losses': train_losses_v2,
    'perplexities': perplexities_v2,
    'hyperparameters': {
        'vocab_size': VOCAB_SIZE,
        'embed_dim': EMBED_DIM,
        'num_heads': NUM_HEADS,
        'num_layers': NUM_LAYERS,
        'dropout': DROPOUT_RATE,
        'learning_rate': LEARNING_RATE,
        'num_epochs': NUM_EPOCHS_V2,
    },
}, 'improved_mini_gpt_checkpoint.pt')

print("Saved: improved_mini_gpt_checkpoint.pt")

print("\nREADY TO TEST GENERATION!")
```

```python

print("TESTING IMPROVED MODEL GENERATION")
print("=" * 60)

def generate_text(model, tokenizer, prompt, max_length=50, temperature=1.0):
    model.eval()
    
    input_ids = tokenizer.encode(prompt, return_tensors='pt').to(device)
    generated = input_ids
    
    with torch.no_grad():
        for _ in range(max_length):
            outputs = model(generated)
            next_token_logits = outputs[:, -1, :] / temperature
            probs = F.softmax(next_token_logits, dim=-1)
            next_token = torch.multinomial(probs, num_samples=1)
            generated = torch.cat([generated, next_token], dim=1)
            
            if generated.size(1) >= max_length:
                break
    
    generated_text = tokenizer.decode(generated[0], skip_special_tokens=True)
    return generated_text

prompts = [
    "The cat sat on the",
    "Once upon a time,",
    "Artificial intelligence is",
    "In the year 2025,",
]

print("\nImproved Model Generation (with Dropout):\n")

for prompt in prompts:
    generated = generate_text(model_v2, tokenizer, prompt, max_length=30, temperature=1.0)
    print(f"Prompt: '{prompt}'")
    print(f"Output: {generated}")
    print("-" * 60)

print("\nGeneration test complete!")
```
